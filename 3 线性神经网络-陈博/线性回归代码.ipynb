```python
import numpy as np
import matplotlib.pylab as plt
```

```python
m=1000
x=np.random.random(size=m)
y=88.*x+77.+np.random.normal(0,10,size=m)
X=x.reshape(-1,1)
```

```python
plt.scatter(x,y)
plt.show()
```

```python
def J(theta,X,y):
   return np.sum(y-X.dot(theta)**2)/len(y)
```

```python
def DJ(theta,X,y):
    y_hat=X.dot(theta)
    #差值 1000r 1l
    y_s=y_hat-y
    # X=1000row 1l
    return X.T.dot(y_s)*2/len(y)
```

```python
def graduent_descent(X,y,init_theta,eta):
    theta=init_theta
    while True:
        graduent=DJ(theta,X,y)
        last_theta=theta
        theta=theta-eta*graduent
        if abs(J(theta,X,y)-J(last_theta,X,y))<1e-8:
            break
    return theta
```

```python
X_b=np.hstack([np.ones((len(X),1)),X])
init_theta=np.zeros(X_b.shape[1])
eta=0.01
theta=graduent_descent(X_b,y,init_theta,eta)
```

```python
z=theta[1]*x+theta[0]
plt.scatter(x,y)
plt.scatter(x,z,color='red')
plt.show()
```
